```python
!pip install stable-baselines3[extra] gymnasium matplotlib --quiet

import numpy as np
import gymnasium as gym
from gymnasium import spaces
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.env_checker import check_env

# --- マップ設定 ---
MAP_H, MAP_W = 20, 20
obstacles = [(8, 6), (8, 10), (8, 14), (12, 6), (12, 10), (12, 14), (16, 6), (16, 10), (16, 14)]

class WallAvoidEnv(gym.Env):
    def __init__(self):
        super().__init__()
        self.action_space = spaces.Box(low=np.array([-1., -1.]), high=np.array([1., 1.]), dtype=np.float32)
        self.observation_space = spaces.Box(low=0.0, high=10.0, shape=(15,), dtype=np.float32)

        self.map = np.zeros((MAP_H, MAP_W))
        for y, x in obstacles:
            self.map[y, x] = 1  # 障害物

        self.max_range = 10.0
        self.reset()

    def reset(self, seed=None, options=None):
        self.robot_pos = np.array([17.0, 10.0])
        self.robot_yaw = np.pi / 2  # 北向き（+y方向）
        self.step_count = 0
        return self._get_lidar(), {}

    def step(self, action):
        self.step_count += 1
        forward = float(action[0]) * 0.3  # 少し遅く
        turn = float(action[1]) * 0.3

        self.robot_yaw += turn
        move = np.array([forward * np.cos(self.robot_yaw), forward * np.sin(self.robot_yaw)])
        next_pos = self.robot_pos + move

        x, y = int(next_pos[0]), int(next_pos[1])
        collision = (x < 0 or x >= MAP_W or y < 0 or y >= MAP_H or self.map[y, x] == 1)

        if collision:
            reward = -10.0
            terminated = True
        else:
            reward = move[1]  # y方向前進で報酬
            self.robot_pos = next_pos
            terminated = False

        if self.step_count > 100:
            terminated = True

        return self._get_lidar(), reward, terminated, False, {}

    def _get_lidar(self):
        # ✅ 360度 → 24度刻み → 15本
        angles = np.deg2rad(np.arange(-180, 180, 24))
        obs = []
        for a in angles:
            theta = self.robot_yaw + a
            for r in np.linspace(0.0, self.max_range, 100):
                x = int(self.robot_pos[0] + r * np.cos(theta))
                y = int(self.robot_pos[1] + r * np.sin(theta))
                if not (0 <= x < MAP_W and 0 <= y < MAP_H):
                    obs.append(r)
                    break
                if self.map[y, x] == 1:
                    obs.append(r)
                    break
            else:
                obs.append(self.max_range)
        return np.array(obs, dtype=np.float32)

    def render(self):
        plt.imshow(self.map, cmap='gray_r', origin='lower')
        plt.plot(self.robot_pos[0], self.robot_pos[1], 'bo')
        plt.title("Robot Position")
        plt.grid(True)
        plt.show()

# --- ✅ 環境構築と学習 ---
env = WallAvoidEnv()
check_env(env)
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=100_000)

# --- ✅ テスト ---
obs, _ = env.reset()
rewards = []
for _ in range(50):
    action, _ = model.predict(obs)
    obs, reward, done, _, _ = env.step(action)
    rewards.append(reward)
    if done:
        break

# --- ✅ プロット ---
plt.plot(rewards)
plt.title("報酬の推移（テスト）")
plt.xlabel("ステップ")
plt.ylabel("報酬")
plt.grid(True)
plt.show()

```
